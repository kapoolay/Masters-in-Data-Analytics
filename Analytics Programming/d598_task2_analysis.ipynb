```python
# D598 Task 2 â€“ Python Data Analysis Script

import pandas as pd
import numpy as np

#load data into df
try: 
    df = pd.read_excel("D598 Data Set.xlsx")
    print("File loaded successfully.")
except FileNotFoundError:
    print("Error: THe data file was not found. Please check filename and path.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

df.head()
```

```python
#identify duplicate rows
df_duplicate = df[df.duplicated()]
df_duplicate

#count number of duplicates
num_duplicates = df_duplicate.shape[0]
print(f"Total Duplicates: {num_duplicates}")
```

```python
#count rows with 0 rev
zero_revenue_count = df[df['Total Revenue'] == 0].shape[0]
print(f"Rows with 0 Total Rev: {zero_revenue_count}")

#calculate debt to income ratio (Total Long-term Debt / Total Revenue)
#include error handling for zero values
df['Debt-to-Income Ratio'] = np.where(
    df['Total Revenue'] != 0,
    df['Total Long-term Debt'] / df['Total Revenue'],
    np.nan #not a number'
)
```

```python
#concatenate new ratio to original df and store in a final df
df_final = df.copy()

#display final df
df_final.head()

#save to an excel
df_final.to_excel('Task 2 - Final Analysis.xlsx', index=False)
```

```python
# Visual 3 - Pie Chart of Business Count by State

# Count businesses per state
state_counts = df_final['Business State'].value_counts()

# Plot the top 10 states only (for readability)
top_states = state_counts.head(10)

plt.figure(figsize=(8, 8))
plt.pie(
    top_states,
    labels=top_states.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)

plt.title('Top 10 States by Number of Businesses')
plt.tight_layout()
plt.show()
```

```python
# Visual 3 - Horizontal Bar Chart of Average Debt-to-Equity Ratio by State

# Calculate average debt-to-equity ratio by state
avg_de_by_state = df_final.groupby('Business State')['Debt to Equity'].mean().sort_values()

# Plot
plt.figure(figsize=(10, 8))
avg_de_by_state.plot(kind='barh', color='mediumseagreen')

plt.title('Average Debt-to-Equity Ratio by State')
plt.xlabel('Debt-to-Equity Ratio')
plt.ylabel('Business State')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
