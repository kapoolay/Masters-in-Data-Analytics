```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load dataset
df = pd.read_csv('Health Insurance Dataset.csv')

# clean column names
df.columns = df.columns.str.lower().str.strip()

# clean numeric columns for plotting per eval comments
df['charges'] = pd.to_numeric(df['charges'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['bmi'] = pd.to_numeric(df['bmi'], errors='coerce')

# drop rows with missing values in relevant numeric fields
df = df.dropna(subset=['charges', 'age', 'bmi'])


df.head()
```

```python
# A.1 visual for age using histogram

sns.set(style="whitegrid")  #background
plt.figure(figsize=(8, 5))  #size
sns.histplot(df['age'], bins=15, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
# A.1 visual for bmi using boxplot

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bmi'], color='lightcoral')
plt.title('BMI Boxplot')
plt.xlabel('BMI')
plt.show()
```

```python
# A.1 visual for smoker count using pie chart

smoker_counts = df['smoker'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(smoker_counts.values, labels=smoker_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Smoker Distribution')
plt.legend(title='Smoker')
plt.show()
```

```python
# A.1 visual for region distribuition using countplot

plt.figure(figsize=(8, 5))
sns.countplot(x='region', data=df, hue='region', palette='Set2', legend=False)
plt.title('Number of Policyholders by Region')
plt.xlabel('Region')
plt.ylabel('Count')

plt.show()
```

```python
# A.2 Provide two bivariate visualizations for each variable selected from part A1.

# Age Variable #1
# scatterplot: age vs charges
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='charges', data=df, hue='smoker', palette='coolwarm')
plt.title('Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.legend(title='Smoker')

plt.show()
# This shows how charges vary by age, with smoker status highlighted. Useful to spot patterns like high charges among older smokers.
```

```python
# D5: Write error-free code in either Python or R to run the parametric test and provide the output and the results of all calculations from the parametric statistical test you perform.
# Research Question: Is there a linear relationship between age and insurance charges? 

import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# load/clean
df = pd.read_csv('Health Insurance Dataset.csv')
df.columns = df.columns.str.lower().str.strip()

# drop any rows with missing values for age/charges
data = df[['age', 'charges']].dropna()

# Pearson correlation
r, p = stats.pearsonr(data['age'], data['charges'])

# Pearson results
print("Pearson Correlation Test (Age v Charges)")
print(f"Correlation coefficient (r): {r:.4f}")
print(f"P-value: {p:.6f}")



# plot
plt.figure(figsize=(7, 5))
sns.regplot(x='age', y='charges', data=data, scatter_kws={'alpha': 0.5})
plt.title('Correlation Between Age and Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy import stats
import numpy as np

# load data and clean 
df = pd.read_csv('Health Insurance Dataset.csv')
df.columns = df.columns.str.lower().str.strip()
df = df.dropna(subset=['charges', 'region'])
df['region'] = df['region'].str.lower().str.strip()

# convert charges to numeric
df['charges'] = pd.to_numeric(df['charges'], errors='coerce')

# drop missing/non-numeric charges
df = df.dropna(subset=['charges', 'region'])
df['region'] = df['region'].str.lower().str.strip()

# groups
regions = sorted(df['region'].unique())
groups = [df.loc[df['region'] == r, 'charges'].values for r in regions]

# sample sizes and medians
print("Per-region counts and medians (charges):")
for r, g in zip(regions, groups):
    print(f"{r.title():<10}  n={len(g):>4}  median=${np.median(g):,.2f}")

# KruskalWallis test
H, p = stats.kruskal(*groups, nan_policy='omit')
print("\nKruskalWallis test (charges/region)")
print(f"H statistic: {H:.4f}")
print(f"P-value: {p:.6f}")
```
